```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
from main import DatasetAnalyzer
```

```python
def generate_random_test_data(num_rows=1000, num_numeric_cols=5, num_categorical_cols=3, num_datetime_cols=2):
    random.seed(42)
    np.random.seed(42)
    
    numeric_data = {
        f'num_col_{i}': np.random.uniform(-100, 100, size=num_rows) for i in range(num_numeric_cols)
    }

    categories = ['A', 'B', 'C', 'D', 'E']
    categorical_data = {
        f'cat_col_{i}': np.random.choice(categories, size=num_rows) for i in range(num_categorical_cols)
    }

    date_range_start = datetime.now() - timedelta(days=730)  # 2 years ago
    datetime_data = {
        f'datetime_col_{i}': [date_range_start + timedelta(days=random.randint(0, 730)) for _ in range(num_rows)] 
        for i in range(num_datetime_cols)
    }
    
    data = {**numeric_data, **categorical_data, **datetime_data}
    df = pd.DataFrame(data)
    
    missing_indices = np.random.choice(df.index, size=int(num_rows * 0.1), replace=False)
    for col in df.columns:
        df.loc[missing_indices, col] = np.nan

    df = pd.concat([df, df.sample(frac=0.2)], ignore_index=True)
    
    return df

# Generate a random dataset
random_df = generate_random_test_data()
```

```python
analyzer = DatasetAnalyzer(random_df)
analyzer.analyze()
```
